```python
# Import pandas for data manipulation and Faker for generating fake data
import pandas as pd
from faker import Faker
import random

# Initialize Faker for generating realistic data
fake = Faker()

# Number of resume records to generate
num_resumes = 100

# Generate resume data
resume_data = []
for _ in range(num_resumes):
    resume_data.append({
        'Candidate_ID': fake.unique.random_int(min=1000, max=9999),
        'Name': fake.name(),
        'Email': fake.email(),
        'Phone_Number': fake.phone_number(),
        'Years_Experience': random.randint(0, 20),
        'Skills': ', '.join(random.sample(['Python', 'SQL', 'Machine Learning', 'Data Analysis', 'Project Management', 'Cloud Computing', 'Web Development'], k=random.randint(2, 5))),
        'Education_Level': random.choice(['High School', 'Associate', 'Bachelor', 'Master', 'PhD']),
        'Last_Job_Title': fake.job(),
        'Expected_Salary_USD': random.randint(40000, 150000),
        'Date_Applied': fake.date_between(start_date='-2y', end_date='today').strftime('%Y-%m-%d')
    })

# Create a DataFrame from the generated data
resumes_df = pd.DataFrame(resume_data)

# Save the DataFrame to a CSV file
resumes_df.to_csv('resumes.csv', index=False)

print('resumes.csv created successfully.')
# Display the first 5 rows of the generated DataFrame
display(resumes_df.head())
```

```python
# Number of employee attrition records to generate
num_employees = 200

# Generate employee attrition data
attrition_data = []
for i in range(num_employees):
    attrition = random.choice([0, 1]) # 0 for no attrition, 1 for attrition
    attrition_data.append({
        'Employee_ID': 1000 + i,
        'Age': random.randint(22, 60),
        'Department': random.choice(['Sales', 'Research & Development', 'Human Resources', 'Marketing', 'Finance']),
        'Job_Role': fake.job(),
        'Monthly_Income': random.randint(2500, 15000),
        'Years_At_Company': random.randint(0, 20),
        'Years_In_Current_Role': random.randint(0, 10),
        'Performance_Rating': random.choice([1, 2, 3, 4]), # 1-Poor, 2-Fair, 3-Good, 4-Excellent
        'Work_Life_Balance': random.choice([1, 2, 3, 4]),
        'Attrition': attrition,
        'Reason_For_Leaving': random.choice(['Better Opportunity', 'Workload', 'Relocation', 'Personal Reasons', 'N/A']) if attrition == 1 else 'N/A'
    })

# Create a DataFrame from the generated data
attrition_df = pd.DataFrame(attrition_data)

# Save the DataFrame to a CSV file
attrition_df.to_csv('employee_attrition.csv', index=False)

print('employee_attrition.csv created successfully.')
# Display the first 5 rows of the generated DataFrame
display(attrition_df.head())
```

```python
import pandas as pd

# Load the employee performance CSV
try:
    performance_df = pd.read_csv('employee_performance.csv')
    print("employee_performance.csv loaded successfully.")
    file_loaded = True
except FileNotFoundError:
    print("Error: employee_performance.csv not found. Please ensure the file is in the correct directory.")
    performance_df = pd.DataFrame() # Initialize an empty DataFrame to prevent NameError
    file_loaded = False

# Proceed with analysis only if the file was loaded successfully and the DataFrame is not empty
if file_loaded and not performance_df.empty:
    # Display basic information and first few rows
    print("\nPerformance Data Info:")
    performance_df.info()
    print("\nFirst 5 rows of Performance Data:")
    display(performance_df.head())

    # Calculate overall average performance metrics
    overall_avg_performance_score = performance_df['Performance_Score'].mean()
    overall_avg_goals_achieved = performance_df['Goals_Achieved_Percentage'].mean()
    overall_avg_feedback_score = performance_df['Feedback_Score'].mean()

    print(f"\nOverall Average Performance Score: {overall_avg_performance_score:.2f}")
    print(f"Overall Average Goals Achieved (%): {overall_avg_goals_achieved:.2f}")
    print(f"Overall Average Feedback Score: {overall_avg_feedback_score:.2f}")

    # Identify Top Performers (e.g., Performance_Score >= 4 and Feedback_Score >= 4 and Goals_Achieved_Percentage >= 90)
    top_performers = performance_df[
        (performance_df['Performance_Score'] >= 4) &
        (performance_df['Feedback_Score'] >= 4) &
        (performance_df['Goals_Achieved_Percentage'] >= 90)
    ]

    print(f"\nNumber of Top Performers identified: {len(top_performers)}")
    if not top_performers.empty:
        print("Top 5 Top Performers:")
        display(top_performers.head())
    else:
        print("No top performers met the criteria.")

    # Identify Employees Needing Improvement (e.g., Performance_Score <= 2 or Feedback_Score <= 2 or Goals_Achieved_Percentage <= 60)
    needs_improvement = performance_df[
        (performance_df['Performance_Score'] <= 2) |
        (performance_df['Feedback_Score'] <= 2) |
        (performance_df['Goals_Achieved_Percentage'] <= 60)
    ]

    print(f"\nNumber of Employees Needing Improvement identified: {len(needs_improvement)}")
    if not needs_improvement.empty:
        print("Top 5 Employees Needing Improvement:")
        display(needs_improvement.head())
    else:
        print("No employees identified as needing improvement based on the criteria.")
else:
    print("Performance data not loaded or is empty, skipping analysis.")
```

```python
def filter_resumes_by_skills(df, required_skills):
    """
    Filters a DataFrame of resumes based on a list of required skills.

    Args:
        df (pd.DataFrame): The DataFrame containing resume data.
        required_skills (list): A list of skills that candidates must possess.

    Returns:
        pd.DataFrame: A DataFrame containing only the resumes that match all required skills.
    """
    # Ensure skills are case-insensitive and handle potential extra spaces
    df_skills_lower = df['Skills'].str.lower().str.strip()

    # Convert required skills to lowercase for case-insensitive matching
    required_skills_lower = [skill.lower() for skill in required_skills]

    # Create a boolean mask to filter rows
    mask = df_skills_lower.apply(lambda x: all(skill in x for skill in required_skills_lower))

    return df[mask]

# Example usage:
# Define the required skills
required_skills = ['Python', 'SQL', 'Machine Learning']

# Filter the resumes_df using the function
filtered_resumes_df = filter_resumes_by_skills(resumes_df, required_skills)

print(f"Original number of resumes: {len(resumes_df)}")
print(f"Number of resumes matching skills {required_skills}: {len(filtered_resumes_df)}")

# Display the first few rows of the filtered DataFrame
display(filtered_resumes_df.head())
```
